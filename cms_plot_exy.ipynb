```python
import numpy as np
import matplotlib.pyplot as plt
import ROOT,root_numpy
import sys
import mplhep as hep
```

```python
inFile = ROOT.TFile.Open("/eos/user/e/efe/tmva/MC13TeV_2017_TAToTTQ_MA-scan_TuneCP5_13TeV_G2HDM-rtc04.root","READ")
```

```python
tree = inFile.Get("TreeInput")
data = tree.AsMatrix(["t_scan_mass"])
h, bins = np.histogram(data, bins=1000)
plt.style.use([hep.style.CMS, hep.style.firamath])
hep.set_style("CMS")
f, axs = plt.subplots()
hep.cms.text("Simulation")
hep.cms.label(loc=0)
hep.histplot(h,bins)
axs.set_xlabel("Scan extra-Higgs mass [GeV]")
plt.tight_layout()
plt.show()
f.savefig("/eos/user/e/efe/SWAN_projects/cms_plot/scan_mass.pdf")
```
